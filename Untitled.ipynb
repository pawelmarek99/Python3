```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Wczytanie danych
tips = sns.load_dataset("tips")

# Wyświetlenie kilku pierwszych wierszy danych
tips.head()
```

```python
# Podstawowe statystyki danych
summary = tips.describe()
print(summary)

# Sprawdzenie wartości null
null_values = tips.isnull().sum()
print("\nWartości null:")
print(null_values)
```

```python
# Wykres rozkładu napiwków
plt.figure(figsize=(8, 6))
sns.histplot(tips['tip'], bins=20, kde=True)
plt.title('Rozkład napiwków')
plt.xlabel('Napiwki')
plt.ylabel('Liczba wystąpień')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxenplot(x='sex', y='total_bill', data=tips)
plt.title('Wydatki w zależności od płci')
plt.xlabel('Płeć')
plt.ylabel('Suma rachunku [$]')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxenplot(x='day', y='tip', data=tips)
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Rozkład napiwków w zależności od dnia tygodnia')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Napiwki')
plt.show()
```

```python
# relacja między wydatkami a napiwkami
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Korelacja między wydatkami a napiwkami')
plt.xlabel('Suma rachunku [$]')
plt.ylabel('Napiwki')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.pairplot(tips, kind='reg', diag_kind='kde')
plt.suptitle('Współczynnik korelacji między wydatkami a napiwkami', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='size', y='total_bill', data=tips)
plt.title('Średnie wydatki w zależności od liczby osób')
plt.xlabel('Liczba osób')
plt.ylabel('Średnia suma rachunku [$]')
plt.show()
```

```python

tips['time'] = tips['time'].astype('category').cat.reorder_categories(['Lunch', 'Dinner'], ordered=True)

plt.figure(figsize=(10, 6))
sns.regplot(x='time', y='total_bill', data=tips)
plt.title('Zależność między czasem trwania posiłku a sumą rachunku')
plt.xlabel('Czas trwania posiłku')
plt.ylabel('Suma rachunku [$]')
plt.show()
```

```python
sns.set_style('whitegrid')

# Przypisanie wartości numerycznych dla kategorii czasu trwania posiłku
time_mapping = {'Lunch': 0, 'Dinner': 1}
tips['time_numeric'] = tips['time'].map(time_mapping)

plt.figure(figsize=(12, 8))
# Używanie sns.regplot() z parametrem scatter_kws do zmiany rozmiaru punktów
sns.regplot(x='time_numeric', y='total_bill', data=tips, scatter_kws={'s': 100})
plt.title('Zależność między czasem trwania posiłku a sumą rachunku')
plt.xlabel('Czas trwania posiłku')
plt.ylabel('Suma rachunku [$]')
plt.xticks([0, 1], ['Lunch', 'Dinner'])  # Ustawienie etykiet na osi X
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
plt.title('Rozkład wydatków w zależności od dnia tygodnia i palenia')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Suma rachunku [$]')
plt.show()
```
